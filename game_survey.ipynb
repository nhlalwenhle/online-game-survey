Psychological Reasons Why Some People Play Video Games

Psychologists and sociologists are only now beginning to understand why the human ability to play is so powerful. But unlocking the mystery behind this desire may do more than help us understand our obsession – it could reshape and improve society in powerful ways. (1)

Gaming does have elements of escapism. The players enter an alternate world, often more fantastical or appealing than their own, and without the ‘glass ceiling’ that seems to constrain real-life action. (2)

Scott Rigby’s research company Immersyve concluded that gaming perfectly targets three psychological needs. (2)
    1. Competency- We like to feel like we are good at something and we like to be recognised for it.
    2. Autonomy-  We want to feel like we are in control of our actions and situations. However, this isn’t always easy in real life.
    3. Relatedness- We like to feel like we matter to other people and like we make a difference within our group or society.

Introduction

The objective of this report is to determine whether people in Umuzi play online games, reason(s) behind why they play, what type of games they play and what type of  games they like.


Interviews

A total of four interviews were conducted and all four were conducted at Umuzi in person. Interview participants were asked a total of 4 questions (Appendix A).
 
Only 3 of the 4 participants indicated that they play online games. Our interviewees stated that they play online games for various reasons such as “ it helps me take my mind off things”, “ it relaxes me”, “its an easy way to connect with people” and “their fun and interesting”. 

One inerviewee stated that she does not play online games because of her schedule, she claimed that she “just doesnt have time to spend playing games”. 

Insights gathered from the interview have lead me to conclude that the majority of people at Umuzi play games when they have time or are trying to relax and unwind.

Online Game Survey. (fig1)

An online (google form) game survey was conducted and sent to Umuzi recruits via email to gather data on whether recurits play, enjoy and how the general experience ofplaying online games.
The online survey asked the participants to play four online games: 
    1. Helicopter game.
    2. Slither. 
    3. Addicting games. 
    4. Paperclips


The participants are then asked to answer a few questions (Appendix B) regarding the experience. The survey was sent to 108 participants but only nine (9) participants completed the survey. Of the nine (9) participants who completed the survey, 77.78% of the participants said they liked playing the online games whilethe other 22.2% of participants didnt enjoy playing the games. (fig 2)
  "Which game was your favourite ?" question informed me that recruits favoured the Helicopter game as their favourite with a score of 44.4% followedby Slither with a score of 33.3% with Paperclips and Addicting games tied for the third spot with 11.1% each. Fig 3.

"Which game was your least favourite ?" question informed me that recruits chose the Paperclips game as their least favourite with a score of 33.3% followed by a tie between Slither and the Helicopter with a score of 22.2% and thereafter Addicting games tied for the third spot with a Missing value with 11.1% each. The missing value is a result of a participant leaving their choice empty. Fig 3.

"Which game did you struggle with ?" gave the participants a chance to show which game they struggled with the most. Slither and the Helicopter game  tied for first place by providing the most amount of struggle for the participants with 33.3% a piece. Their followed by Addicting games with 22.2% and paperclips provided the least resistance with 11.1%. fig 4.

“Would you recommend any of the games to other people?”, this was a “YES”or “NO” question where 66.7% of the participant chose YES and only 33.3% chose “NO” . Fig 5


Observations
While observing the survey participants play the game,the following insights were gained:
    • The lack of playability on cellphones frustrated participants.
    • Slither’s mouse controls frustrated some players.
    • Lack of instructions on some games frustrated the participants.

Conclusion
Given the data collected and analysed, we can safely assume that recruits in Umuzi do play online games and thus introducing games with a bit more variety and a bit more action such as the Helicopter game which was chosen as the most favourite game within the set of games provided.
